```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from scipy.stats import uniform as sp_uniform
lgbmr = lgb.LGBMRegressor(random_state = 1, n_jobs = -1)

params = {'n_estimators' : sp_randint(400, 402),
         'max_depth' : sp_randint(1, 20),
         'learning_rate' : sp_uniform(0.01, 0.1)}

lgbm_hpt = RandomizedSearchCV(lgbmr, param_distributions = params,
                                 cv = 5, n_iter = 100, n_jobs = -1, random_state = 0)

lgbm_hpt.fit(X, y)
```
